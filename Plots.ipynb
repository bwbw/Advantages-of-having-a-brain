```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data from the file
def read_txt(filename):
    with open(filename, 'r') as file:
        return [float(line.strip()) for line in file]

# Plot the smoothed data
def smooth(window_size, filename, length=None, la=None):
    data = read_txt(filename)[:length]
    moving_avg = np.convolve(data, np.ones(window_size)/window_size, mode='valid')
    plt.plot(range(window_size-1, len(data)), moving_avg, label = la)
    plt.legend(loc='upper left')

def calculate_mean_std(filenames, length=None):
    means = []
    stds = []

    for filename in filenames:
        # Delete the first 200 data points
        data = read_txt(filename)[200:length]
        # Calculate mean and standard deviations
        means.append(np.mean(data))
        stds.append(np.std(data))
    return means, stds
```

```python
# test on mutiple runs (Average Fitness vs. Generation)

f1 = ['data/avg_fitness_data_c50m20n20.txt',
      'data/avg_fitness_data_c50m20n20 a.txt',
      'data/avg_fitness_data_c50m20n20 b.txt',
      'data/avg_fitness_data_c50m20n20 c.txt',
      'data/avg_fitness_data_c50m20n20 d.txt']
l1 = ['Run 1', 'Run 2', 'Run 3', 'Run 4', 'Run 5']
x1 = [1, 2, 3, 4, 5]

for i in range(len(f1)):
    smooth(200, f1[i], length=1000, la=l1[i])
plt.xlabel('Generations')
plt.ylabel('Average fitness')
plt.title('Average fitness of the bugs over generations')
plt.show()

# mean and standard deviation
plt.errorbar(x1, calculate_mean_std(f1)[0], yerr=calculate_mean_std(f1)[1], fmt='-o', color='black', ecolor='r')
plt.xlabel('Runs')
plt.ylabel('Mean average fitness')
plt.title('Error bar plot of mean average fitness on different runs')
plt.show()

print(calculate_mean_std(f1))
```
